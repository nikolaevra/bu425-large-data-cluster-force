```python
# Library imports
import pandas as pd
import seaborn as sns
sns.set(context="notebook", palette="Spectral", style = 'darkgrid', font_scale = 1.5, color_codes=True)
```

```python
merchant = pd.read_csv("/home/nikolaevra/datasets/elo/merchants.csv")
```

```python
for column in merchant.columns:
    num_missing = merchant[[column]].isna().sum()[0]
    if num_missing > 0:
        print("{} {}".format(column, num_missing))
```

```python
cols_to_show = [
    'avg_sales_lag3',
    'avg_sales_lag6',
    'avg_sales_lag12',
    'avg_purchases_lag3',
    'avg_purchases_lag6',
    'avg_purchases_lag12',
    'active_months_lag3',
    'active_months_lag6',
    'active_months_lag12',
]
pplot = sns.pairplot(merchant[cols_to_show].dropna().sample(5000))
pplot.savefig("pair_plot_merchants.png")
```
